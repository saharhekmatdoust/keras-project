# Build a Regression Model in Keras
In this project, I will build a regression model using the deep learning Keras library, and then I will experiment with increasing the number of training epochs and changing number of hidden layers and we will see how changing these parameters impacts the performance of the model.
In this project,we will usethe  data about concrete compressive strength.


the data can be found here again: https://cocl.us/concrete_data. To recap, the predictors in the data of concrete strength include:

Cement

Blast Furnace Slag

Fly Ash

Water

Superplasticizer

Coarse Aggregate

Fine Aggregate

```python
import numpy as np
import pandas as pd
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

##### checking whether data has null values or not

```python
concrete_data.isnull().sum()
```

### Deviding dataframe into 2 data frames as predictors(X) and target (y)

```python
X=concrete_data.drop(columns='Strength',axis=1)
X.head()
```

```python
y=concrete_data['Strength']
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
n_columns=X.shape[1]
n_columns
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```


###  Building a model with One hidden layer of 10 nodes, and a ReLU activation function and Using the adam optimizer and the mean squared error  as the loss function.

```python
# define regression model
def regression_model1():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_columns,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
MSE=[]
```

```python
for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=None)
    model1 = regression_model1()
    model1.fit(X_train,y_train,epochs=50, verbose=2)
    prediction=model1.predict(X_test)
    MSE.append(mean_squared_error(y_test,prediction))
    print(MSE)
    
```

```python
import statistics

print("Mean is {}\nStandard deviation is {}".format(statistics.mean(MSE),statistics.stdev(MSE)))
```


### Repeating with a normalized version of the data

```python
X_norm=(X-X.mean())/X.std()
X_norm.head()
```

```python
MSEB=[]
for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(X_norm,y,test_size=0.3,random_state=None)
    model1 = regression_model1()
    model1.fit(X_train,y_train,epochs=50, verbose=2)
    prediction=model1.predict(X_test)
    MSEB.append(mean_squared_error(y_test,prediction))
    print(i,MSEB)
    
```

```python
print("Mean is {}\nStandard deviation is {}".format(statistics.mean(MSEB),statistics.stdev(MSEB)))
```


### Repeating the process but using 100 epochs this time for training.

```python
MSEC=[]
for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(X_norm,y,test_size=0.3,random_state=None)
    model1 = regression_model1()
    model1.fit(X_train,y_train,epochs=100, verbose=2)
    prediction=model1.predict(X_test)
    MSEC.append(mean_squared_error(y_test,prediction))
    print(i,MSEC)
    
```

```python
print("Mean is {}\nStandard deviation is {}".format(statistics.mean(MSEC),statistics.stdev(MSEC)))
```


### Using a neural network with the following instead:Three hidden layers, each of 10 nodes and ReLU activation function.

```python
# define regression model
def regression_model2():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_columns,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))

    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
MSED=[]
for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(X_norm,y,test_size=0.3,random_state=None)
    model2 = regression_model2()
    model2.fit(X_train,y_train,epochs=50, verbose=2)
    prediction=model2.predict(X_test)
    MSED.append(mean_squared_error(y_test,prediction))
    print(i,MSED)
    
```

```python
print("Mean is {}\nStandard deviation is {}".format(statistics.mean(MSED),statistics.stdev(MSED)))
```

```python
result={'Method':['one hidden layer/50 epochs','1 hidden layer/50 epochs/normalized data',
                 '1 hidden layer/100 epochs/normalized data','3 hidden layers/50 epochs/normalized data']
       ,'mean of MSE':[statistics.mean(MSE),statistics.mean(MSEB),statistics.mean(MSEC),statistics.mean(MSED)],
      'Standard deviation of MSE':[statistics.stdev(MSE),statistics.stdev(MSEB),statistics.stdev(MSEC),statistics.stdev(MSED)]
      }
results=pd.DataFrame(result)
results
```

# Results and Conclusion:
### mean of MSE with normalized data has higher value compared with unnormalized data.
### Standard deviation of MSE of normalized data is much smaller than unnormalized data.
### increasing the number of epochs results in considerable drop in both mean and std of MSE.
### increasing the number of hidden layers to 3 results in the best performance among 4 methods.
